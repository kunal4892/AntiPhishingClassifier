```python
cleanedData.head()
dataCSVCLeaned = cleanedData[["isIp","urlLen","isredirect","haveDash","domainLen","nosOfSubdomain"]].values

isIPInURL = dataCSVCLeaned[:,0]
urlLength = dataCSVCLeaned[:,1]
hasRedirect = dataCSVCLeaned[:,2]
haveDashInURL = dataCSVCLeaned[:,3]
domainLength = dataCSVCLeaned[:,4]
noOfSubDomain = dataCSVCLeaned[:,5]

isIPInURLZero = isIPInURL[np.where(noOfLabels == 0)]
print("isIPInURL for legitimate website -> Mean is : " ,np.mean(isIPInURLZero))
print("isIPInURL for legitimate website -> Standard Deviation is : ", np.std(isIPInURLZero))

isIPInURLOne = isIPInURL[np.where(noOfLabels == 1)]
print("isIPInURL for phishing website -> Mean is : " ,np.mean(isIPInURLOne))
print("isIPInURL for phishing website -> Standard Deviation is : ", np.std(isIPInURLOne))

print("\n")

urlLengthZero = urlLength[np.where(noOfLabels == 0)]
print("urlLength for legitimate website -> Mean is : " ,np.mean(urlLengthZero))
print("urlLength for legitimate website -> Standard Deviation is : ", np.std(urlLengthZero))

urlLengthOne = urlLength[np.where(noOfLabels == 1)]
print("urlLength for phishing website -> Mean is : " ,np.mean(urlLengthOne))
print("urlLength for phishing website -> Standard Deviation is : ", np.std(urlLengthOne))

print("\n")

hasRedirectZero = hasRedirect[np.where(noOfLabels == 0)]
print("hasRedirectZero for legitimate website -> Mean is : " ,np.mean(hasRedirectZero))
print("hasRedirectZero for legitimate website -> Standard Deviation is : ", np.std(hasRedirectZero))

hasRedirectOne = hasRedirect[np.where(noOfLabels == 1)]
print("hasRedirect for phishing website -> Mean is : " ,np.mean(hasRedirectOne))
print("hasRedirect for phishing website -> Standard Deviation is : ", np.std(hasRedirectOne))

print("\n")

haveDashInURLZero = haveDashInURL[np.where(noOfLabels == 0)]
print("haveDashInURL for legitimate website -> Mean is : " ,np.mean(haveDashInURLZero))
print("haveDashInURL for legitimate website -> Standard Deviation is : ", np.std(haveDashInURLZero))

haveDashInURLOne = haveDashInURL[np.where(noOfLabels == 1)]
print("haveDashInURL for phishing website -> Mean is : " ,np.mean(haveDashInURLOne))
print("haveDashInURL for phishing website -> Standard Deviation is : ", np.std(haveDashInURLOne))

print("\n")

domainLengthZero = domainLength[np.where(noOfLabels == 0)]
print("domainLength for legitimate website -> Mean is : " ,np.mean(domainLengthZero))
print("domainLength for legitimate website -> Standard Deviation is : ", np.std(domainLengthZero))

domainLengthOne = domainLength[np.where(noOfLabels == 1)]
print("domainLength for phishing website -> Mean is : " ,np.mean(domainLengthOne))
print("domainLength for phishing website -> Standard Deviation is : ", np.std(domainLengthOne))

print("\n")

noOfSubDomainZero = noOfSubDomain[np.where(noOfLabels == 0)]
print("noOfSubDomain for legitimate website -> Mean is : " ,np.mean(noOfSubDomainZero))
print("noOfSubDomain for legitimate website -> Standard Deviation is : ", np.std(noOfSubDomainZero))

noOfSubDomainOne = noOfSubDomain[np.where(noOfLabels == 1)]
print("noOfSubDomain for phishing website -> Mean is : " ,np.mean(noOfSubDomainOne))
print("noOfSubDomain for phishing website -> Standard Deviation is : ", np.std(noOfSubDomainOne))

print("\n")
```
